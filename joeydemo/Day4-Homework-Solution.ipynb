```python
import pandas as pd
sales = pd.read_csv('sales.csv', header=0, index_col='id')
# display(sales)
# print(sales.color.value_counts())

from sklearn import preprocessing as pp
cols = ['cost', 'sales']
sales[cols] = sales[cols].apply(lambda x : x.fillna(x.mean()))
sales[cols] = sales[['cost', 'sales']].apply(lambda x : pp.scale(x, with_mean = True, with_std = True))
#display(sales)
sales['colorcode'] = pd.Categorical(sales.color).codes
#display(sales)
dummies = pd.get_dummies(sales.color, drop_first = False)
sales = pd.concat([sales, dummies], axis = 1)
display(sales)
```

```python
import pandas as pd
sales = pd.read_csv('sales.csv', header=0, index_col='id')

from sklearn import preprocessing as pp
import pandasql as ps
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

coloravgsql = 'select color, avg(cost) as avgcost, avg(sales) as avgsales from sales group by color'
coloravgdf = pysqldf(coloravgsql)
#display(coloravgdf)

joinsql = '''select id, s.color, coalesce(s.cost, a.avgcost) as cost, s.cost, s.sales
, coalesce(s.sales, a.avgsales) as sales
from sales as s
join coloravgdf as a on s.color = a.color
where s.cost is null or s.sales is null
order by s.color'''
display(pysqldf(joinsql))
```
